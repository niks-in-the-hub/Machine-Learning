```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
```

```python
# The IRIS dataset is a very clasically researched dataset in the machine learning field.
#The notebook is intended to build functions that use a model stacking approach on IRIS.
#The same can be used as a general helper function on any regular dataset.
df=pd.read_csv(r"C:\Users\Nikkitha\Downloads\IRIS.csv")
```

```python
#mapping labels to integers

df=df.replace('Iris-setosa',0)  #indexing Iris-setosa as 0
df=df.replace('Iris-versicolor',1)  #indexing Iris-versicolor as 1
df=df.replace('Iris-virginica',2)  #indexing Iris-virginica as 2
```

```python
df.sample(10)  #checking a sample for replication
```

```python
df.sample(frac=1)  #shuffling the dataset
```

```python
X=df.iloc[:,:-1]  #extracting feature vectors to X
```

```python
len(df.columns)
```

```python
Y=df.iloc[:,4]  #extracting labels 
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
def VotingMethod(X_train,Y_train):
    clf1=LogisticRegression(solver='lbfgs', multi_class='multinomial')  #clf1=LR()
    clf2=RandomForestClassifier(n_estimators=50)                         #clf2=RFC()
    clf3=GaussianNB()                                                    #clf3=GNB()
    eclf1=VotingClassifier(estimators=[  ('lr',clf1), ('rf',clf2), ('gnb',clf3)  ],
                       voting='hard')
    print ("Feature matrix shape" , X_train.shape)  
    print ("Label matrix shape", Y_train.shape)
    eclf1=eclf1.fit(X_train, Y_train)
    Y_pred=eclf1.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    print("Accuracy of the Voting Classifier is:", acc)
    return Y_pred, acc, eclf1
```

```python
VotingMethod(X_train,Y_train)
```
